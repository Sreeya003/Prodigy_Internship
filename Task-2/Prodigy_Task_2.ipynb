```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(style="white", color_codes=True)
import warnings
warnings.filterwarnings("ignore")
```

```python
# Load the Titanic dataset
url = "https://web.stanford.edu/class/archive/cs/cs109/cs109.1166/stuff/titanic.csv"
titanic_data = pd.read_csv(url)
titanic_data.head()
```

```python
titanic_data.duplicated()
```

```python
titanic_data.isna().sum()
```

```python
# Fill missing values in 'Age' with the median age
titanic_data['Age'].fillna(titanic_data['Age'].median(), inplace=True)
```

```python
print(titanic_data.shape)
```

```python
print(titanic_data.describe())
```

## Task 2- Exploratory Data Analysis

```python
sns.countplot(data = titanic_data,x='Survived',palette=['#ff0000','#26cc23'])

plt.title('Survival Count on the Titanic')
plt.xlabel('Survived')
plt.ylabel('Count')
plt.show()
```

```python
sns.countplot(x='Pclass', data=titanic_data)
plt.title('Pclass Distribution')
plt.show()
```

```python
# Violin plot of Age by Survived
plt.figure(figsize=(10, 6))
sns.violinplot(x='Survived', y='Age', data=titanic_data)
plt.title('Age Distribution by Survival Status')
plt.show()
```

```python
# Pair plot of selected features
sns.pairplot(titanic_data[['Age', 'Fare', 'Pclass', 'Survived']], hue='Survived')
plt.show()
```

```python
titanic_data.drop(columns=['Name'], inplace=True)

titanic_data.drop(columns=['Sex'], inplace=True)
```

```python
# Correlation matrix heatmap
corr_matrix = titanic_data.corr()
plt.figure(figsize=(12, 8))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()
```
